```python
# Create a Scatter plot indicating the salary against age
plt.figure(figsize=(10, 6))

sns.scatterplot(data=data, x='age', y='salary', color = "blue",
                edgecolor = 'linen', alpha = 0.5)
plt.title('Age vs Salary', fontweight = 'bold')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

Observation


we observ that in general , there is a relationshio between age and salary of emp.
age increases , the salary also increased, though ther are some outasliers present as well.

```python
# Create the scatter plot of the  healthy_eating and active_lifestyle.

plt.figure(figsize=(10,8))

sns.scatterplot(data=data, x= 'healthy_eating', y ='active_lifestyle',
                color = 'red', edgecolor = 'linen', alpha= 0.5)

plt.title("Healthy_Eating  vs active_lifestyle", fontweight = 'bold')
plt.xlabel("healthy_eating")
plt.ylabel("active_lifestyle")
plt.show()
```

Observation

we observ that in general , there is a relationship between healthy_eating and active_lifestyle of employee, healthy eating increases , the active lifestyle also increases.

```python
# Create the scatter plot agains the sal and healthy eating
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='healthy_eating', y='salary', color='black', edgecolor='linen', alpha=0.7)
plt.title('Healthy Eating vs Salary', fontweight='bold')
plt.xlabel('Healthy Eating Score')
plt.ylabel('Salary')
plt.show()
```

Observation

as we found that the healthy eating increases the salary also increases.

```python
# Create a count plot for the groups features

plt.figure(figsize=(10,6))

sns.countplot(data=data, x= 'groups', hue='groups', edgecolor='linen', alpha=0.7)

plt.title("Count plots of groups", fontweight='bold')
plt.xlabel("groups")
plt.ylabel("Count")
plt.show()
```

Observation


we see that most employees are belongs to the group A and O, while group A has the most no of employee.

```python
# Create histogram about salary featues

plt.figure(figsize= (10,6))

sns.histplot(data=data, x = "salary",
             color= "yellow" ,edgecolor = "black", bins = 10, )
plt.title("salary", fontweight = "bold")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

Observation


We see that the salaries are not uniform , which is not neccessarily discrepency
 but we do observe there are employee with slareies on either extremities of the problem

 Based on th histogram we can use that the count of highest earning employee is very less.
